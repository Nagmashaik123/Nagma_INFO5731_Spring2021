```python
# Importing requests module for feeding wb url to return the response object
import requests
import csv
from bs4 import BeautifulSoup

# Iterating through for loop to extract top 100 abstracts
Y =[]
X =[item*10 for item in range(1,11)]

for item in X:
  URL = 'https://citeseerx.ist.psu.edu/search?q=natural+language+processing&t=doc&sort=rlv&start={}'.format(item)
  page = requests.get(URL)
  soup = BeautifulSoup(page.content, 'html.parser')
  for item in soup.find_all("div",{"class":"snippet"}):
    Y.append(item.text)
```

# **Question 2**

(30 points). Write a python program to **clean the text data** you collected above and save the data in a new column in the csv file. The data cleaning steps include:

(1) Remove noise, such as special characters and punctuations.

(2) Remove numbers.

(3) Remove stopwords by using the [stopwords list](https://gist.github.com/sebleier/554280).

(4) Lowercase all texts

(5) Stemming. 

(6) Lemmatization.

```python
import nltk
nltk.download()
```

```python
import re
import nltk
nltk.download('wordnet')
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from textblob import Word

# Removing punctuations and special characters
df['Noise_Removal'] = df['Abstract'].str.replace('[^\w\s]','')
df['Noise_Removal'] = df['Noise_Removal'].apply(lambda item: ''.join(re.sub(r"[^a-zA-Z0-9]+", ' ', char) for char in item ))

# Removing digits
df['Digit_Removal'] = df['Noise_Removal'].apply(lambda x: ''.join([item for item in x if not item.isdigit()]))

# Stop words removal
stopwords = stopwords.words('english')
df['Stopwords_Removal'] = df['Digit_Removal'].apply(lambda item: " ".join(item for item in item.split() if item not in stopwords))

# To lower Case comversion
df['To_Lower_Case'] = df['Stopwords_Removal'].apply(lambda item: " ".join(item.lower() for item in item.split()))

df


```

```python
# Stemming
st = PorterStemmer()
df['Stemming'] = df['To_Lower_Case'].apply(lambda x: " ".join([st.stem(X) for X in item]))

# Lemmatization
df['Lemmatization'] = df['Stemming'].apply(lambda x: " ".join([Word(word).lemmatize() for word in x.split()]))

```

```python
# Saving data to CSV from data frame
df.to_csv('ProcessedAbstract.csv', index=False)
```

```python
# Parts of Speech processing
from nltk.tokenize import word_tokenize
posTagging = []
for stmt in df['To_Lower_Case']:
  word = word_tokenize(stmt)
  posTagging.append(nltk.pos_tag(word))
posTagging


```

```python
# Counts of Parts of Speech
from collections import Counter
count_array = []
for tags in posTagging:
  counts = Counter( tag for word,  tag in tags)
  count_array.append(counts)
count_array
```
