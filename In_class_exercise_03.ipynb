```python
# Importing the modules for web scraping
from bs4 import BeautifulSoup
import requests
import re

# web url to scrap the data from
url = 'https://www.oneindia.com/elections/members-of-parliament-list/'

# making a request for the webpage using get function which return python response object .i.e HTML view
response = requests.get(url)

#Step3: Parse the page

# Parsing the data using Beautifulsoap for reading webcontent which provides readable libraries
soup = BeautifulSoup(response.text, 'lxml')
soup.prettify()

# Get the tr elements based on tr tag and read the td elements for the respective table row.
parliament = []

# finds all the table row based on tr tag declaration
table_rows = soup.find_all('tr') 

#Iterate the row abd read the text assosciated with each table row.
for tr in table_rows:
    td = tr.find_all('td')
    row = [i.text for i in td]
    parliament.append(row)  

#Refining and printing the unempty furnished list
parliamentList = [i for i in parliament if i]
count=0

finalList=[]
#Iterating through the list and storing it through dictionary so the title hearders could be appended
for items in parliamentList:
  candidateName = items[3].replace("\n","")
  partyName = items[4].replace("\n","")
  data = {"serialNo"  : items[0],
         "stateName" : items[1],
         "constituency" : items[2],
         "candidateName": candidateName,
         "partyName"   : partyName
         }
  finalList.append(data)

for item in finalList:
  print(item['serialNo'], ',',item['stateName'], ',' ,item['constituency'], ',' ,item['candidateName'], ',' ,item['partyName'])

  

    


```

```python
#Creating a csv file and storing all the data to the file

import csv
filename = 'Members_Of_Parliament.csv'
with open(filename, 'w', newline='') as f: 
    w = csv.DictWriter(f,['serialNo','stateName','constituency','candidateName','partyName']) 
    w.writeheader() 
    for item in finalList: 
        w.writerow(item) 
```
